### Encoder Data

```python
df_covid['Location_Level'] = df_covid['Location_Level'].astype('category')
# Assigning numerical values and storing in another column
df_covid['Location_Level'] = df_covid['Location_Level'].cat.codes

df_covid['Location'] = df_covid['Location'].astype('category')
# Assigning numerical values and storing in another column
df_covid['Location'] = df_covid['Location'].cat.codes

df_covid['Province'] = df_covid['Province'].astype('category')
# Assigning numerical values and storing in another column
df_covid['Province'] = df_covid['Province'].cat.codes

df_covid['Country'] = df_covid['Country'].astype('category')
# Assigning numerical values and storing in another column
df_covid['Country'] = df_covid['Country'].cat.codes

df_covid['Island'] = df_covid['Island'].astype('category')
# Assigning numerical values and storing in another column
df_covid['Island'] = df_covid['Island'].cat.codes

df_covid['Special_Status'] = df_covid['Special_Status'].astype('category')
# Assigning numerical values and storing in another column
df_covid['Special_Status'] = df_covid['Special_Status'].cat.codes

df_covid['Time_Zone'] = df_covid['Time_Zone'].astype('category')
# Assigning numerical values and storing in another column
df_covid['Time_Zone'] = df_covid['Time_Zone'].cat.codes
```

```python
df_covid
```

### Replace Missing Value

```python
# Replace Missing Value Province
Province = df_covid['Province']
df_covid.Province = df_covid.Province.fillna(value=df_covid.Province.mean())

# Replace Missing Value Island
Island = df_covid['Island']
df_covid.Island = df_covid.Island.fillna(value=df_covid.Island.mean())

# Repalce Missing Value Time_zone
Time_zone = df_covid['Time_Zone']
df_covid.Time_Zone = df_covid.Time_Zone.fillna(value=df_covid.Time_Zone.mean())

# Replace Missing Value Special_Status
Special_Status = df_covid['Special_Status']
df_covid.Special_Status = df_covid.Special_Status.fillna(value=df_covid.Special_Status.mean())

# Replace Missing Value Special Total_Cities
Total_Cities = df_covid['Total_Cities']
df_covid.Total_Cities = df_covid.Total_Cities.fillna(value=df_covid.Total_Cities.mean())

# Replace Missing Value Total_Urban_Villages
Total_Urban_Villages = df_covid['Total_Urban_Villages']
df_covid.Total_Urban_Villages = df_covid.Total_Urban_Villages.fillna(value=df_covid.Total_Urban_Villages.mean())

# Replace Missing Value Total_Rural_Village
Total_Rural_Villages = df_covid['Total_Rural_Villages']
df_covid.Total_Rural_Villages = df_covid.Total_Rural_Villages.fillna(value=df_covid.Total_Rural_Villages.mean())

# Replace Missing Value Growth_Factor_of_New_Cases
Growth_Factor_of_New_Cases = df_covid['Growth_Factor_of_New_Cases']
df_covid.Growth_Factor_of_New_Cases = df_covid.Growth_Factor_of_New_Cases.fillna(value=df_covid.Growth_Factor_of_New_Cases.mean())

# Replace Missing Value Growth_Factor_of_New_Deaths
Growth_Factor_of_New_Deaths = df_covid['Growth_Factor_of_New_Deaths']
df_covid.Growth_Factor_of_New_Deaths = df_covid.Growth_Factor_of_New_Deaths.fillna(value=df_covid.Growth_Factor_of_New_Deaths.mean())
```
